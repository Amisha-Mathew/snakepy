```python
# Description and InvoiceNo play no significant role in the prediction model
# Hence Dropping  Description and InvoiceNo

df.drop('Description', axis=1, inplace=True)
df.drop('InvoiceNo', axis=1, inplace=True)
```

```python
# To ensure secured dropping of columns
# printing all the column names
list(df.columns.values)
```

```python
# checking if null values are present
df.isnull().sum()
```

```python
# Dropping tuples where the customerID is null
# Becomes irrelevant to the Prediction Model

df.dropna(subset=['CustomerID'], inplace = True)
df.isnull().sum()
```

```python
# Checking the datatypes of columns
print(df.dtypes)
```

```python
# Inserting a column to determine the total purchase 

df['Total'] = df['Quantity'] * df['UnitPrice']
df.describe()
```

```python
# Printing all the unique records of a column
# This concludes that the Country is irrelevant to the prediction model
countrylist = df['Country'].unique()

countrydf = pd.DataFrame(countrylist, columns =['countryName'])
countrydf
```

```python
# Generating CodeNumber for Countries
# This makes country-based predictions a lot easier

# converting type of columns to 'category'
df['Country'] = df['Country'].astype('category')

# Assigning numerical values and storing in another column
df['codeNo'] = df['Country'].cat.codes

df.head(30)
```

```python
# Checking if null values are present
# More focussed towards ensuring codeNo is filled in properly
df.isnull().sum()
```

```python
# Checking for one parameter 
# whether the codeNo has been correctly mapped or not
count = 0

if ((df['Country']=='United Kingdom') & (df['codeNo'] != 35)).any():
    count += 1
    
print(count)
```

```python
# TODO
# checking for all 'country' parameter rightly mapped
checker_df = df.loc[(df['Country']=='France') & (df['codeNo'] != 13)]
checker_df.head(10)
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
#target variable creation
training_df=df[(df['InvoiceDate'] >= pd.Timestamp(2010,12,2)) & (df['InvoiceDate'] <= pd.Timestamp(2011,8,31))]

training_df.head()
```

```python
not_churn_df=df[(df['InvoiceDate'] >= pd.Timestamp(2011,9,1)) & (df['InvoiceDate'] <= pd.Timestamp(2011,12,31))]

not_churn_df.head()
```

```python
cust_id= training_df.CustomerID.unique()
cust_id
```

```python
not_churn_id= not_churn_df.CustomerID.unique()
not_churn_id
```

```python
churn_id=[]
for i in cust_id:
    if i not in not_churn_id:
        churn_id.append(i)
```

```python
df=df.assign(Churn=[1 if x in churn_id else 0 for x in df['CustomerID']])
df.head()
```

```python
df=df.drop(['StockCode','codeNo','InvoiceDate', 'CustomerID'],axis=1)
```

```python
df.head()
```

```python
df=df.dropna(axis=1)
df_with_sep_countries= pd.get_dummies(df)
df_with_sep_countries
```

```python
df_without=df_with_sep_countries.drop('Churn', axis=1)

df_without
```

```python
#Decision Tree using ID3 algorithm

from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_without,df_with_sep_countries['Churn'], test_size=0.10)
```

```python
# Create Decision Tree classifer object
dt_classifier = DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifer
dt_classifier = dt_classifier.fit(X_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = dt_classifier.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
